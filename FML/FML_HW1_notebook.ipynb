```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
data = np.genfromtxt('FML/housingUnits.csv', delimiter=',')
```

```python
#1)
print("correlation of predictor variable 2 and outcome:")
print(np.corrcoef(data[1:,1],data[1:,7])[1,0])
print("correlation of predictor variable 3 and outcome:")
print(np.corrcoef(data[1:,2],data[1:,7])[1,0])
```

```python
print("R^2 of linear regression fit to predictor variable 2 and outcome:")
X = data[1:,1].reshape(-1, 1)
y = data[1:,7].reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

model = LinearRegression().fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
#new table replacing data with normalized data
normalized_data = np.copy(data)
normalized_data[1:,1] = data[1:,1]/data[1:,3]
normalized_data[1:,2] = data[1:,2]/data[1:,3]
```

```python
#3)
print("simple linear regression using predictor variable 1:")
X = normalized_data[1:,0].reshape(-1, 1)
y = data[1:,7].reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
model = LinearRegression().fit(X_train,y_train)
print(model.score(X_test,y_test))

plt.scatter(X_test, y_test, s=1)
y_hat = model.predict(X_test)
plt.plot(X_test, y_hat,color='k')
plt.text(40, 10000, f"R^2 = {model.score(X_test,y_test):.3f}")
plt.xlabel('Median age of the houses in the block (in years)')
plt.ylabel('Median house value in the block (in dollars)')
plt.title('Median age of the houses in the block against median house value in the block')
plt.show()
```

```python
#4)
print("multiple linear regression:")
X = normalized_data[1:,:6]

y = data[1:,7].reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
model = LinearRegression().fit(X_train,y_train)
print(model.score(X_test,y_test))
y_hat = model.predict(X)
plt.scatter(y, y_hat-y, s=1)
plt.xlabel('Median house value in the block') 
plt.ylabel('Residual from multiple linear regression')  
plt.title('Graph of residuals')
plt.show()
```
